DATA AUGMENTATION

```python
from PIL import Image, ImageEnhance, ImageOps
import os

image_path="premium_photo-1664474619075-644dd191935f.jpeg"
image=Image.open(image_path)
```

```python
augmented_dir="augmented_images"
os.makedirs(augmented_dir,exist_ok=True)
```

```python
def augment_image(image,output_dir):
  flipped=ImageOps.mirror(image)
  flipped.save(os.path.join(output_dir,"flipped.jpg"))

  rotated=image.rotate(38)
  rotated.save(os.path.join(output_dir,"rotated.jpg"))
  enhancer=ImageEnhance.Brightness(image)
  enhanced=enhancer.enhance(1.5)
  enhanced.save(os.path.join(output_dir,"enhanced.jpg"))
  contrast=ImageEnhance.Contrast(image).enhance(2.0)
  contrast.save(os.path.join(output_dir,"contrast.jpg"))
  greyscale=ImageOps.grayscale(image)
  greyscale.save(os.path.join(output_dir,"greyscale.jpg"))
```

```python
augment_image(image,augmented_dir)
print("augmented images save in:",augmented_dir)
print("generated files",os.listdir(augmented_dir))
```

```python
augmented_dir = "augmented_images"

augmented_images = os.listdir(augmented_dir)

plt.figure(figsize=(12, 8))

for i, img_file in enumerate(augmented_images):
    img_path = os.path.join(augmented_dir, img_file)
    img = Image.open(img_path)

    plt.subplot(2, 3, i + 1)

    plt.imshow(img, cmap="gray" if img.mode == "L" else None)
#L stands for "Luminance", which is a grayscale image mode.
# Use grayscale for grayscale images
plt.title(img_file.split('.')[0])  # Use file name as title
plt.axis("off")  # Hide axes

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "image.png"
image = cv2.imread(image_path)

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray
plt.imshow(gray,cmap=('gray'))
```

```python
gray
```

```python
low_pass=cv2.GaussianBlur(gray,(15,15),0)
low_pass
```

```python
plt.imshow(low_pass,cmap="gray")
```

```python
# Sharpen the image using a kernel
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])
sharpened = cv2.filter2D(gray, -1, sharpen_kernel)
plt.figure(figsize=(12, 12))
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```

```python
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)  # X direction
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)  # Y direction

plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')
```

```python
edges=cv2.Canny(gray,100,200)
edges
plt.imshow(edges,cmap="gray")
```

```python
median_filtered=cv2.medianBlur(gray,5)
median_filtered
plt.imshow(median_filtered,cmap="gray")
```
